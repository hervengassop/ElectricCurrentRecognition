# Dataset Analysis

Notebook to analyze and visualize all electric currents (in the time domain) from dataset.

## Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Leitura do Dataset

```python
dataset = pd.read_csv('ElectricCurrentDataset.txt')
dataset.head()
```

### Classes do dataset

```python
classes=[]
for i in dataset['Device']:
    if i not in classes:
        classes.append(i)
print('Exemplos\t Classe\n')
for classe in classes:
    num_exemplos = dataset[dataset['Device']==classe].count()[0]
    print(num_exemplos,'\t\t',classe)
```

### Conjunto de Dados

```python
for classe in classes:
    sinal = dataset[dataset['Device']==classe]
    sinal.drop(['Device'], axis=1, inplace=True)
    for i in range(len(dataset)):
        if dataset['Device'].loc[i] == classe:
            plt.plot(sinal.loc[i])
    plt.title(classe)
    plt.show()
```
